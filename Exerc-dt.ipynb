#Exercício 1 

Um estabelecimento quer implantar um sistema de controle de estoque, mas o dono do estabelecimento possui praticamente zero experiencia com sistema de computador. Para começar pediu para que os funcionários comecem a contar o estoque de materiais utilizados e produtos vendidos. No entanto, ele não especificou um formato padrão de dados nem quando, então cada funcionário foi colocando os dados do seu jeito. No fim do mês, o dono percebeu que fazer um relatório seria muito trabalhoso, mas ele teve a ideia de pedir para o sobrinho (você) dele, pois sabia que ele tinha experiencias com computador e é muito inteligente pois é aluno da FEA USP e membro da FEA DEV :> .

O dono fez alguns pedidos para o sistema:

• mostre resultados semanais
• mostre resultado mensal
• arrume a data para formato datetime

Faça um sistema que mostre o resultado do mês.

```python
#Dados

data = ["2 may 24", "05-03-24", "05/08 2024", "may 10 2024", "05-13-2024", "05-16-2024" ,"05-17-2024", "2024 may 24", "28/05/24", "30/05/24", "may 31 2024"]

dados = {
    "materiais" : np.random.randint(10, 1000, 11),
    "produtos" : np.random.randint(10, 1000, 11)
}

dados = pd.DataFrame(dados, index=data)
dados
```

```python
#Uma das resposta possível

dados.index = pd.to_datetime(dados.index)

#Resultado semanal

resul_sem = dados.resample('W').sum()

#Resultado mensal

resul_men = dados.resample('M').sum()

print(resul_sem)
print('------------------------------------')
print(resul_men)
```

#Exercício 2

Agora que o dono tem um pouco mais de noção sobre computação, ele pediu um sistema mais complexo, e agora tambêm ele tem um banco de dados maior.

Atenda os novos critêrios:

•Crie um novo DataFrame com relatório de somas das vendas de cada mês
•Crie um nova coluna no mesmo DataFrame mostrando variação entre mês atual e anterior
•Mostre a média semanal de vendas de cada mês
•Mostre resultado de últimos dois meses
•Faça o mesmo mas agora semanal
•Faça uma função que traga resultados da semana inserido

```python
#Dados

data2 = pd.date_range('06-01-2024', periods=30, freq='3B')

dados2 = {
     "materiais" : np.random.randint(10, 1000, 30),
     "produtos" : np.random.randint(10, 1000, 30)   
}

dados2 = pd.DataFrame(dados2, index=data2)
dados2
```

```python
#Uma das possíveis respostas

#Relatório mensal
relatorio_men = dados2.resample('M').sum()
relatorio_men_var = relatorio_men.diff()

relatorio_men['var_materiais'] = relatorio_men_var['materiais']
relatorio_men['var_produtos'] = relatorio_men_var['produtos']

relatorio_men_med = dados2.resample('M').mean()

relatorio_men['med_materiais'] = relatorio_men_med['materiais']
relatorio_men['med_produtos'] = relatorio_men_med['produtos']

ult_2_mes = relatorio_men[relatorio_men.shape[0] - 2:]


#Relatório semanal

relatorio_sem = dados2.resample('W').sum()
relatorio_sem_var = relatorio_sem.diff()

relatorio_sem['var_materiais'] = relatorio_sem_var['materiais']
relatorio_sem['var_produtos'] = relatorio_sem_var['produtos']

relatorio_sem_med = dados2.resample('W').mean()

relatorio_sem['med_materiais'] = relatorio_sem_med['materiais']
relatorio_sem['med_produtos'] = relatorio_sem_med['produtos']

def fun_relatorio_sem(data):
    data = pd.to_datetime(data)
    semana = data.week

    for i in relatorio_sem.index:

        if i.week == semana:

            print(relatorio_sem.loc[i])

            break

fun_relatorio_sem('23 August 2024')
```
